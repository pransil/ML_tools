```python
# Build a 3 dimensional array data set as follows:
# Dimension 0: random floats, identically distributed over [0, 1], for 1000 values
# Dimension 1: 1000 floats , with 4 clusters of equal size  where
# the first cluster has a mean of 0.2 and a standard deviation of 0.05
# the second cluster has a mean of 0.5 and a standard deviation of 0.05
# the third cluster has a mean of 0.8 and a standard deviation of 0.05
# Dimension 2 has 4 groups where
# group 0 has 300 values with an iid distribuion over [0, 0.28]
# group 1 has 300 values with an iid distribuion over [0.3, 0.58]
# group 2 has 200 values with an iid distribuion over [0.6, 0.78]
# group 3 has 200 values with an iid distribuion over [0.8, 1]

import numpy as np
import matplotlib.pyplot as plt


# Build a numpy array of 1000 floats, identically distributed over [0, 1]
data0 = np.random.rand(1000)
# Print the shape of data0
print(data0.shape)

# Build a numpy array of 1000 floats, with 4 clusters of equal size
# where the first cluster has a mean of 0.2 and a standard deviation of 0.05
# the second cluster has a mean of 0.5 and a standard deviation of 0.05
# the third cluster has a mean of 0.8 and a standard deviation of 0.05
data1 = np.random.normal(loc=0.2, scale=0.05, size=300)
data2 = np.random.normal(loc=0.5, scale=0.05, size=300)
data3 = np.random.normal(loc=0.8, scale=0.05, size=400)
# Concatenate data1, data2, data3 into a single array
data4 = np.concatenate((data1, data2, data3))
print(data4.shape)

# Build a numpy array of 1000 floats, with 3 groups where
# group 0 has 300 values with an iid distribuion over [0, 0.28]
# group 1 has 300 values with an iid distribuion over [0.3, 0.58]
# group 2 has 200 values with an iid distribuion over [0.6, 0.78]
# group 3 has 200 values with an iid distribuion over [0.8, 1]

data5 = np.random.rand(300)
# Scale to between 0 and 0.28
data5 = data5 * 0.28

data6 = np.random.rand(300)
# Scale to between 0.3 and 0.58
data6 = data6 * 0.28 + 0.3

data7 = np.random.rand(200)
# Scale to between [0.6, 0.78]
data7 = data7 * 0.18 + 0.6

data8 = np.random.rand(200)
# Scale to between [0.8, 1]
data8 = data8 * 0.18 + 0.8

# Concatenate data5, data6, data7 into a single array
data9 = np.concatenate((data5, data6, data7, data8))
print(data9.shape)






```

```python
# Combine data0, data4, data8 into a 3 dimensional array
data10 = np.array([data0, data4, data9])

# Print the shape of data9
print(data10.shape)

# Plot data10[0] vs data10[1]
plt.scatter(data10[0], data10[1])
```

```python
# Plot data10[1] vs data10[2]
plt.scatter(data10[1], data10[2])
```

```python
# Plot data10[0] vs data10[2]
plt.scatter(data10[0], data10[2])
```

```python
# Reformat data9 into an array of 1000 3 dimensional points
data10 = data9.reshape(1000, 3)
print(data10.shape)
print(data10[:5])
```
